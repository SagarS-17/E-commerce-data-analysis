```python
import pandas as pd
import numpy as np
```

```python
# Import csv file as dataframe

df = pd.read_csv('messy_ecommerce_sales_data.csv')
```

```python
# Dropping duplicate rows

df[df.duplicated()]
df.drop_duplicates(inplace = True)
```

```python
# Stripping white spaces from column titles

df.columns
df.columns = df.columns.str.strip()
```

```python
# Dropping incorrect values from date column 

df = df[df['Order_Date'] != 'abc']

# Dropping dates with only year value available

df = df[df['Order_Date'].str.len() >= 9 ]

# Converting date values to numeric format 

df = df.replace('Jan 5 2023','05-01-2023')

# converting date format from dd/mm/yy to dd-mm-yyyy

df['Order_Date'] = df['Order_Date'].str.replace('/', '-', regex = False)
df['Order_Date'] = pd.to_datetime(df['Order_Date']).dt.strftime('%d-%m-%Y')
```

```python
# Cleaning and standardising product category column

df['Category'] = np.where(df['Product'] == 'Basketball', 'Sports', df['Category'])
df['Category'] = np.where(df['Product'] == 'Biography', 'Books', df['Category'])
df['Category'] = np.where(df['Product'] == 'Blender', 'Home', df['Category'])
df['Product'] = np.where(df['Product'] == 'Headphones', 'Headphone', df['Product'])
df['Category'] = np.where(df['Product'] == 'Headphone', 'Electronics', df['Category'])
df['Category'] = np.where(df['Product'] == 'Jeans', 'Clothing', df['Category'])
df['Category'] = np.where(df['Product'] == 'Lamp', 'Home', df['Category'])
df['Category'] = np.where(df['Product'] == 'Laptop', 'Electronics', df['Category'])
df['Category'] = np.where(df['Product'] == 'Microwave', 'Home', df['Category'])
df['Category'] = np.where(df['Product'] == 'Shoes', 'Clothing', df['Category'])
df['Product'] = np.where(df['Product'] == 'Smartphones', 'Smartphone', df['Product'])
df['Category'] = np.where(df['Product'] == 'Smartphone', 'Electronics', df['Category'])
df['Category'] = np.where(df['Product'] == 'Smartwatch', 'Electronics', df['Category'])
df['Category'] = np.where(df['Product'] == 'T-shirt', 'Clothing', df['Category'])
df['Category'] = np.where(df['Product'] == 'Vacuum', 'Home', df['Category'])
df['Category'] = np.where(df['Product'] == 'Yoga Mat', 'Sports', df['Category'])
```

```python
# Cleaning quantity column and dropping null values

df['Quantity'] = df['Quantity'].str.replace('a', '', regex = False)
df['Quantity'] = df['Quantity'].str.replace('-', '', regex = False)
df = df.dropna(subset = ['Quantity'])

# Converting quantity column to float type

df['Quantity'] = df['Quantity'].astype(float)
df['Quantity'].value_counts()
```

```python
# Cleaning price column and dropping null values

df = df.dropna(subset = ['Price'])
df['Price'] = df['Price'].str.replace('-', '', regex = False)
df['Price'] = df['Price'].str.replace('$', '', regex = False)
df['Price'] = df['Price'].str.replace('four hundred', '400', regex = False)
df = df.drop(df[df['Price'] == 'abd'].index)

# Converting price column to float type

df['Price'] = df['Price'].astype(float)
```

```python
# Cleaning total column and filling empty columns with product of price and quantity

df['Total'] = df['Total'].abs()
df['Total'] = df['Total'].fillna(df['Price'] * df['Quantity'])
```

```python
# Creating new csv file from dataframe

df.to_csv('cleaned_data.csv', index = False)
```
